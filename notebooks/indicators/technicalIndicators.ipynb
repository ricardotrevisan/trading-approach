```python
import yfinance as yf
import pandas as pd

from ta import add_all_ta_features
from ta.utils import dropna

tick = yf.Ticker(str('PETR4.SA'))
hist_data = tick.history(period="max")  

hist_data.head()
```

```python
import matplotlib.pyplot as plt 

# Substantiate data with momentum indicators
mom_data = add_all_ta_features(hist_data, open="Open", high="High", low="Low", close="Close", volume="Volume") 

corr = mom_data.corr()

f = plt.figure(figsize=(40,40))
plt.matshow(corr, fignum=f.number)
plt.xticks(range(mom_data.shape[1]), mom_data.columns, fontsize=14, rotation=45)
plt.yticks(range(mom_data.shape[1]), mom_data.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation on Technical Indicators', fontsize=16)
```
