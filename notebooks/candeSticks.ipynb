```python
import pandas as pd
from pandas_datareader import data as web
import plotly.graph_objects as go

stock = 'PETR4.SA'
df = web.DataReader(stock, data_source='yahoo', start='01-01-2020')
avg30 = df.Close.rolling(window=30, min_periods=1).mean()
avg50 = df.Close.rolling(window=50, min_periods=1).mean()
df.head()
```

```python
estrovenga = {
    'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low,
    'type': 'candlestick',
    'name': stock,
    'showlegend': True
}

estrovenga30 = {
    'x': df.index,
    'y': avg30,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'blue'
            },
    'name': 'Moving Average of 30 periods'
}

estrovenga50 = {
    'x': df.index,
    'y': avg50,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'red'
            },
    'name': 'Moving Average of 30 periods'
}

data = [estrovenga, estrovenga30, estrovenga50]

layout = go.Layout({
    'title': {
        'text': 'Trading Analysis',
        'font': {
            'size': 15
        }
    }
})

fig = go.Figure(data=data, layout=layout)
fig.show()
```
